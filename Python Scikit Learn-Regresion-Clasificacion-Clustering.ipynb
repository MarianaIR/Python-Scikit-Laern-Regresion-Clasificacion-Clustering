## BASE DE ENTRENAMINETO Y BASE DE TEST

Separacion en Base de Entrenamiento y en Base de Test, para la construccion de un MODELO, es importante hacer esto, ya que se usa la base de entrenamiento y los datos de la base de entrenamiento para contruir el modelo, sin embargo se necesita garantizar que el modelo funciona con datos con los cuales NO construi el modelo, entonces para eso tengo la base de test, es decir construyo el modelo con lo que tengo en la base de entrenamiento y voy a evaluar el modelo con los datos que tengo en la base test


```python

# Escribimos FUNCION con la cual vamos hacer la separacion entre la base de ENTRENAMIENTO y la
# base de TEST, ejecutando e importando la funcion -->
from sklearn.model_selection import train_test_split
```

```python
# La Funcion de arriba va a crear 4 Parametros que van a ser 4 base de datos y la vamos a llamar-->

###!!!! x_entrenamiento, x_test, y_entrenamiento, y_test = train_test_split(x, y) !!! ###

# Donde Y, es la variable DEPENDIENTE, variable categoria, respuesta.
# Donde X, es la variable INDEPENDIENTE, datos de entrada independiente, con los que voy a intentar
# PREDECIR el dato o la categoria que tengo que en dato Y
# CON LO QUE TENGO EN LA VARIABLE X VOY A INTENTAR PREDECIR LO QUE TENGO EN LA VARIABLE Y
```

```python

#Creamos las Variables X , Y -->
x = base_de_credito[["edad", "vr_prestamo"]]
y = base_de_credito["status_prestamo"] # En status_prestamos, 0 = NO esta atrasado y 1 = esta atrasado
```

```python
# Hay otro PARAMETRO para tener en cuenta que es el tamaño de la base de TEST que yo quiero, generalmente
# tiene entre el 25% y 30% de la base y la base de ENTRENAMIENTO tiene entre el 70% y 75% del tamaño
# Por parametro , por defecto la Funcion viene con el 25% de la base de TEST, pero por motivos didacticos vamos
# dejar en un 30% (test_size=0.3)--->

x_entrenamiento, x_test, y_entrenamiento, y_test = train_test_split(x, y, test_size=0.3) # ANTES DE EJECUTAR ESTE
# CODIGO TENEMOS QUE CREAR LAS VARIABLES (X, Y)
```

```python
#Antes de seguir con lo de arriba , debemos pasar los valores de la columnas que aparecen con
# TEXTO, ya que un ARBOL DE DECISION NO trabaja con texto, transformamos los texto en valores -->
x.head()
```

## ARBOL DE DECISION
Construccion de un modelo de ARBOL DE DECISION, ya separamos nuestra base de datos en base de ENTRENAMIENTO y base de TEST como asi tambien ya hicimos el ajuste o tratamiento de los datos erroneos en nuestra base de datos

```python
# 1°) Importamos las bibliotecas que vamos a necesitar -->
from sklearn.tree import DecisionTreeClassifier
# 2°) Importamos funcion para ver cual es la calidad de nuestro modelo, para calcular
# la Precision de los resultados -->
from sklearn.metrics import accuracy_score
# 3°) Para realizar graficos de nuestro ARBOL de Decision-->
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 4°) Creamos el MODELO -->
dtc = DecisionTreeClassifier(max_depth=3) # dtc= variable, max_depth= tamaños de lineas de nuestro arbol
```

```python
# 5°) Ajuste del MODELO, contruimos el modelo con la BASE DE ENTRENAMIENTO, con su base de
# INDEPENDIENTE y DEPENDIENTE de X y Y -->
dtc.fit(x_entrenamiento, y_entrenamiento)
# NOS DEVUELVE PARAMETROS DEL MODELO
```

```python
# 6°) Hacemos PREDICCIONES del MODELO, ponemos a prueba con la BASE DE TEST -->
predicciones = dtc.predict(x_test)
predicciones
```

```python
# 7°) Vemos cual es el ACCURACY SCORE de nuestro MODELO -->
accuracy_score(y_test, predicciones) # y_test= datos reales de salida
# NOS DEVUELVE VALORES CORRECTO (87 %), BUEN RESULTADO!
```

```python
# 8°) Vemos el ARBOL DE DECISION graficamente, creamos una lista con los nombres que tenemos
# en las variables que tenemos el el ARBOL -->
columnas = x.columns
fig, axes = plt.subplots(nrows=1, ncols=1, figsize = (20, 20)) # Funcion de la biblioteca de MATPLOTLIB
tree.plot_tree(dtc, feature_names=columnas, class_names=["0", "1"], filled=True);
```

**Sera que podemos PREDECIR los casos que tendran atraso en el prestamo (morosidad)??

RTA= con el 0.879(88%) de PRECISION podemos predecir en el futuro los casos que tendran atraso con los prestamos


**Que variables tenemos en nuestro ARBOL que el ARBOL decidio??

-El decidio PRESTAMO como % porcentaje del INGRESO, SI, el <= 30% lo clasifica como NO, si tiene mas de 30% clasifica como SI

-SI, la TASA DE INTERES es <=14%, lo clasifica como NO

-SI, la clasificacion de RIESGO DE LA PERSONA <=0, la clsifica como moroso

-Los casos que clasifica como ATRASO(MOROSO), SI, si la persona en ALQUILER DE VIVIENDA, tiene mas posibilidad de atraso

**SI creamos mas linea en el ARBOL(max_depth=4), puede que suba la calidad de PREDCCION , en este caso sube  0.892(89%), lo unico que puede demorar mas en la prediccion de nuestro MODELO!!!

## REGRESION
Un modelo de Regresion es una ecuacion matematica que genera una relacion lineal entre Variables, entre la variable INDEP. y DEPEND., la regresion evalua cuanto cambia una variable como resultado de la variacion de otra variable.
Ejemplo, Cambio de precio de una vivienda dado a que aunmenta su tamaño en metros cuadrados.

## REGRESION LINEAL
El objetivo de la Regresion Lineal, es entonces la Linea que mejor se ajuste a los Datos

*DOS MODELOS DE TIPO DE REGRECION LINEAL*

### 1-REGRESION LINEAL SIMPLE
Dispone de una ecuacion con una Variable Dependiente y una Variable Independiente.
### 2-REGRESION LINEAL MULTIPLE
Dispone de una ecuacion con una Variable Dependiente y MAS de una Variable Independiente(mas de una variable de entrada)



*Por lo general un MODELO LINEAL SIMPLE, tiene mejor resultado que un MODELO LINEAL MULTIPLE ya que estamos trabajando con mas variables*

## CLUSTERING
CONSTRUCCION DE LA BASE

```python
# 1°) Creamos un nuevo DATAFRAME de la base de datos que subimos(clientes_centro_comercial)-->
df_cluster = pd.read_csv("clientes_centro_comercial.csv", sep=";")
df_cluster.head()
```

```python
# 2°) Nuestro objetivo con la base es crear Grupos de clientes que tengan caracteristicas similares dentro de ellas
# y cada grupo a su vez tenga caracteristicas diferentes, en este MODELO DE CLUSTERING NO tenemos una variable respuesta
# NO hay variable DEPENDIENTE, entonces NO es necesario separar la informacion en base de TEST y base ENTRENAMIENTO, porque
# NO estamos tratando de llegar a un resultado especifico, si no llegar a un grupo que sea el mas adecuado para la creacion
# de grupos de clientes con caracteristicas especificas.
# Analizamos la base de datos y vemos si hay datos que esten faltando y el tamaño de nuestra base de datos --->
df_cluster.info()
# tenemos 200 lineas, 5 colum, y NO tenemos mas datos NULOS, tenemso 4 variables numericas y 1 variable de texto(object)
```

```python
# 3°) Vamos a estar trabajando con 2 variables, la variable INGRESO y la variable SCORE_GASTOS, construimos
# el modelo con estas 2 variables, para poder visualizarlas de mejor manera creamos un nuevo DATAFRAME, con estas
# 2 variables que vamos a estar usando -->
x = df_cluster.iloc[:,[3,4]] # Iloc para visualizar elementos de la base con su ubicacion
x.head()
```

```python
# 4°) Comprobamos cuales son las escalas o rangos de las variables, para eso realizamos 2 graficos de Histograma o
# Distribucion para ver cual es el formato, la estructura que tienen, usamos la biblioteca SEABORN -->
sns.displot(x["ingreso"]) # la escala va desde 2000 hasta 14000
```

```python
# Luego de ver la diferencia en rangos de escala necesitamos ESTANDARIZAR para que nuestras
# variables sean comparables en sus escalas, para esto vamos a importar una biblioteca
# Esta FUNCION transforma cada una de las columnas que tenemos en la base , en este caso 2
# Variables y las va a dejar a cada una de ellas con una MEDIA de 0 y una DESVIACION ESTANDAR de 1,
# de esta forma van a ser comparables -->
from sklearn.preprocessing import StandardScaler
```

```python
# 6°) Construimos nuestro ESTANDARIZADOR -->
standard_scaler = StandardScaler()
```

```python
# 7°) Creamos una nueva base que tenga los valores ESTANDARIZADOS -->
x_ss = standard_scaler.fit_transform(x)
```

## K-MEANS

Una forma que podemos medir la calidad de los Grupos y de los Modelos, es calculando la diferencia que esxiste entre el centro del CLUSTER, el CENTROIDE y los puntos que forman el CLUSTER, mientras menos sea esa diferencia es mejos, va a llegar un punto en que la diferencia va a ser 0, en el que vamos a tener un numero de Grupos o de CLUSTER igual al numero de puntos en nuestro modelo

```python
# 1°) Importamos la biblioteca K-MEANS -->
from sklearn.cluster import KMeans
```

```python
# 2°) Luego creamos una LISTA -->
wcss = []
```

```python
# 3°) Creamos un LOOP, el objetivo en este LOOP es colocar diferentes numeros de CLUSTER, para ver en que
# momento el CLUSTER de de ser relevante, deja de crear diferencias significativas -->
for i in range(1,11): # desde 1 CLUSTER hasta 10 CLUSTER
    kmeans = KMeans(n_clusters = i, random_state=0) # KMeans = a nuestra funcion KMeans
    kmeans.fit(x_ss)# colocamos la base de datos Estandarizada
    wcss.append(kmeans.inertia_) # colocamos en nuestra lista los resultados
```

```python
# 4°) Colocamos en un grafico nuestra informacion -->
plt.plot(range(1,11),wcss)
plt.xlabel("Numero de Clusters") # eje x
plt.ylabel("Inertia")

#En el grafico podemos ver que cae la INERTIA a medida que aumentamos los CLUSTER
#Este metodo lo llamamos, el METODO del CODO, el obejtivo es ver en que momento se produce el
# quiebre, en este caso tenemos 2, uno en 3 y el otro en 5
# ENTONCES VAMOS A TRABAJAR INICIALMENTE CON 5 CLUSTERS
```

```python
# 5°) Construimos el Modelo con 5 CLUSTERS -->
kmeans = KMeans(n_clusters=5) #KMeanas = funcion que importamos, n_clusters= cantidad de clusters
```

```python
# 6°) Colocamos los CLUSTERS que creamos(5), en la base que importamos y que luego ESTANDARIZAMOS-->
df_cluster["Cluster"] = kmeans.fit_predict(x_ss)# "Cluster" = creamos columna Cluster
```

```python
df_cluster.head()
```

```python
# 7°) Realizamos un garfico con estas informaciones, para ver como tenemos nuestra informacion -->
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 0, 'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 0, 'score_gastos'].values, c='red', label='Cluster 1')
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 1, 'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 1, 'score_gastos'].values, c='yellow', label='Cluster 2')
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 2, 'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 2, 'score_gastos'].values, c='blue', label='Cluster 3')
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 3, 'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 3, 'score_gastos'].values, c='green', label='Cluster 4')
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 4, 'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 4, 'score_gastos'].values, c='brown', label='Cluster 5')

plt.xlabel('Ingreso ($)')
plt.ylabel('Score de Gastos')
plt.legend()
```

-El GRUPO CLUSTER 3, tiene ingresos bajos pero gastan bastante
